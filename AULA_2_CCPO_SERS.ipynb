#Exemplo- **motor de inducao trifasico**

*1. incerção dos dados de placa pelo usuario*

```python
#potencia util do motor= potencia mecanica= kwatts
PU = float(input('insira potencia util do motor em kwatts: '))
#rendimento n = PU / P
n = float(input('insira o rendimento percentual:   ')) / 100
# Fator de potencia  fp = P / S
fp = float(input('insira o fator de potencia (de 0 a 1):   '))
```

```python
PU, n, fp
```

2.calculo das potencias ativa (P), aparente (S) e reativa (Q)

```python
#n = PU / P
#potencia ativa (P)
P = PU / n
#potencia aparente (S)
#fp = P / S
S = P / fp
#potencia reativa (Q)
Q = (S**2 - P**2)**0.5
```

```python
P, S, Q
```

3. APRESENTAÇÃO DOS RESULTADOS

```python
#f-string
#limitar resultado em duas casas decimais
#ex.{p:.2/3/4/5f}

print(f'a potencia ativa sera de {P:.2f} kW')
print(f'a potencia aparente sera de {S:.2f} kVA')
print(f'a potencia reativa sera de {Q:.2f} kVAr')
```
