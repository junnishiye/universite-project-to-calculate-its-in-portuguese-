#Exemplo- **motor de inducao trifasico**

*1. incerção dos dados de placa pelo usuario*

```python
#potencia util do motor= potencia mecanica= watts
PU = int(input('insira potencia util do motor em watts: '))
#rendimento n = PU / P
n = float(input('insira o rendimento (de 0 a 1):   '))
# Fator de potencia  fp = P / S
fp = float(input('insira o fator de potencia (de 0 a 1):   '))
```

```python
PU, n, fp
```

2.calculo das potencias ativa (P), aparente (S) e reativa (Q)

```python
#n = PU / P
#potencia ativa (P)
P = PU / n
#potencia aparente (S)
#fp = P / S
S = P / fp
#potencia reativa (Q)
Q = (S**2 - P**2)**0.5
```

```python
P, S, Q
```

3. APRESENTAÇÃO DOS RESULTADOS

```python
#f-string
print(f'a potencia ativa sera de {int(P)} W')
print(f'a potencia aparente sera de {int(S)} VA')
print(f'a potencia reativa sera de {int(Q)}VAr')
```
